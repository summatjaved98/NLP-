```
from pyspark.ml.feature import (Tokenizer,StopWordsRemover,CountVectorizer,IDF,StringIndexer)
```

```
tokenizer = Tokenizer(inputCol='texts',outputCol='token_text')
stopwordremove = StopWordsRemover(inputCol='token_text',outputCol='stop_token')
count_vec = CountVectorizer(inputCol='stop_token',outputCol='c_vec')
idf = IDF(inputCol='c_vec',outputCol='tf_idf')
ham_spam_tonumeric = StringIndexer(inputCol='classes',outputCol='label')
```

```
from pyspark.ml.feature import VectorAssembler
```

```
clean_up = VectorAssembler(inputCols=['tf_idf','length'],outputCol='features')
```

```
from pyspark.ml.classification import NaiveBayes
```

```
nb = NaiveBayes
```

```
from pyspark.ml import Pipeline
```

```
data_pipe_prep = Pipeline(stages=[ham_spam_tonumeric,tokenizer,stopwordremove,count_vec,idf,clean_up])
```

```
data = data.drop('_c2','_c3','_c4')
data.filter("texts is NULL")
data = data.na.drop()
data.show()
```

```
cleaner = data_pipe_prep.fit(data)
```

```
clean_data = cleaner.transform(data)
```

```
clean_data.columns
```

```
clean_data = clean_data.select('label','features')
```

```
clean_data.show()
```

```
splits = clean_data.randomSplit([0.7, 0.3], 1234)
train = splits[0]
test = splits[1]
```

```
nb = NaiveBayes(smoothing=1.0, modelType="multinomial")
```

```
model = nb.fit(train)
```

```
data.printSchema
```

```
test_result = model.transform(test)
```

```
test_result.show()
```

```
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
```

```
acc_eval = MulticlassClassificationEvaluator()
```

```
acc = acc_eval.evaluate(test_result)
```

```
print("Acc of NB MODEL")
print(acc)
```
